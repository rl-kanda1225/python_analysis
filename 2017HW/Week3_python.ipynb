# 第一題：題目：給定一個矩陣m，裡面的值都大於0。請將矩陣m裡大於等於1的數字都變成1，小於1的變成0。

```python
m = np.array([[0.1,1,0.2],[2,0.6,7]])
for i in range(m.shape[0]):
    for j in range(m.shape[1]):
        if m[i,j]>=1:
            m[i,j]=1
        elif m[i,j]<1:
            m[i,j]=0
print(m)
```
